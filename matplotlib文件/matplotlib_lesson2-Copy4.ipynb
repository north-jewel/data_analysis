# matplotlib 绘制基本图形 画个笑脸

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
var = np.linspace(-np.pi,np.pi,256,endpoint=True)  # 指定x轴范围
```

```python
x=2*np.cos(var)
y=2*np.sin(var)

left_x = 0.5*np.cos(var)-1
left_y=0.5*np.sin(var)+1

right_x = 0.5*np.cos(var)+1
right_y=0.5*np.sin(var)+1
```

```python
 # 1,2,3   --> 1, 2 ,3 ,1,2,3
```

```python
plt.figure(1)
```

```python
plt.title('smile face')  # 添加的位置位置注意一下
plt.plot(x,y,color='red',linewidth=4.0,linestyle='-',label='sin',alpha=0.5)  # label 图例
plt.plot(left_x,left_y,color='red',linewidth=4.0,linestyle='-',label='sin',alpha=0.5)
plt.plot(right_x,right_y,color='red',linewidth=4.0,linestyle='-',label='sin',alpha=0.5)
plt.legend()
```
