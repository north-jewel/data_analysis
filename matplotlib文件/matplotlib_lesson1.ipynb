# matplotlib 绘制基本图形

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi,np.pi,256,endpoint=True)  # 指定x轴范围
```

```python
y = np.sin(x)   # 根据x轴的点 确定对应的y的点  ,然后通过描点法 将图形画出来
```

```python
z = np.cos(x)
```

```python
plt.figure(1)
```

```python
# help(plt.figure)
```

```python
plt.title('sin & cos')  # 添加的位置位置注意一下
plt.plot(x,y)
plt.plot(x,z)
```
