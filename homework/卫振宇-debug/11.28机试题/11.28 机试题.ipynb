```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import *
import pandas as pd
plt.rcParams['font.sans-serif']=['SimHei']  # 用来正常显示中文标签  
plt.rcParams['axes.unicode_minus']=False  # 用来正常显示负号
```

```python
x= np.linspace(0,5,5)
y = [39,8,0,1,6]
fig,ax = plt.subplots(figsize = (12,6))
ax.grid(axis = 'y')
ax.set_title('1802c班违纪情况')
ax.set_ylim(0,45)
ax.set_xticks(x)
ax.set_xticklabels(['未带电脑总数','未带手机总数','未关投影仪次数','未观门窗次数','校查违纪'])
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
label = ['未带电脑总数','未带手机总数','未关投影仪次数','未观门窗次数','校查违纪']
for a,b,c in zip(x,y,label):
    ax.annotate([c,b],xy = (a,b),xycoords = 'data',xytext = (+10,+30),textcoords = 'offset points',
                 arrowprops = dict(arrowstyle = '<-',connectionstyle = 'arc3,rad=.5'))
ax.bar(x,y)
```
