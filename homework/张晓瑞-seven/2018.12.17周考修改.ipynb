# 排量最大的车

```python
df6 = df['Displacement'].apply(lambda x:x.replace('纯电动','').replace('T','').replace('L','').replace('--','1.0').replace(',','').replace("'",'').replace(' ',''))
df666 = df6.apply(lambda x:x[-3:])
df['max'] = df666
df.sort_values(['max'],ascending=False).head(1)
# df6
```
