```python
df2 = df[['省级政区名称','性别']]
df2_re = df2[df['性别'].notnull()]

pt = pd.crosstab(df2['省级政区名称'],df2_re['性别'])
# pt
pt['女性占比'] = (pt['女'] / pt['男'])
# pt
pt2 = pt.sort_values('女性占比',ascending=False)
pt2.head()
# pt2.index[:10]
```

```python
fig = plt.figure(figsize=(8,4))
# pt2.plot.bar()
index = pt2.index[:10]
plt.bar(range(10),pt2['女性占比'][:10],tick_label=index,color='gray')
plt.title('不同省份女性领导占比')
# plt.xticks(range(10),['辽宁省', '陕西省', '吉林省', '山西省', '四川省', '安徽省', '江西省', '山东省', '福建省', '云南省'])
plt.xlabel('省份')
plt.ylabel('比例')
plt.show()
# pt2.index[:10]
```

# 年龄情况,专业情况

```python
data_age = df[['出生年份','党委书记姓名','年份']]
data_re = data_age[data_age['出生年份'].notnull()]
df1 = 2017 - data_re['出生年份']
# df1.head()
df_yearmin = data_re[['党委书记姓名','年份']].groupby(data_re['党委书记姓名']).min()
df2 = df_yearmin['年份'].groupby(df_yearmin['年份']).count()
# df2.head()
df_yearmax = data_re[['党委书记姓名','年份']].groupby(data_re['党委书记姓名']).max()
df3 = df_yearmin['年份'].groupby(df_yearmin['年份']).count()
# df3.head()
```

```python
data_major = df[['党委书记姓名','年份','专业：人文','专业：社科','专业：理工','专业：农科','专业：医科']]
data_major_re = data_major[data_major['专业：人文'].notnull()]
print(data_major_re.head())
print('-'*60)
data_major_re['专业'] = data_major_re[['专业：人文','专业：社科','专业：理工','专业：农科','专业：医科']].idxmax(axis=1)
print(data_major_re.head())
print('-'*60)
data_major_re_st = data_major_re[['专业','党委书记姓名']].drop_duplicates()
print(data_major_re_st.head())
print('-'*60)
df4 = data_major_re_st['专业'].groupby(data_major_re_st['专业']).count()
print(df4.head())
print('-'*60)
df5 = pd.crosstab(data_major_re['年份'],data_major_re['专业'])
print(df5.head())
print('-'*60)
df5['社科比例'] = df5['专业：社科'] / (df5['专业：理工']+df5['专业：医科']+df5['专业：社科']+df5['专业：农科']+df5['专业：人文'])
df5['人文比例'] = df5['专业：人文'] / (df5['专业：理工']+df5['专业：医科']+df5['专业：社科']+df5['专业：农科']+df5['专业：人文'])
df5['理工农医比例'] = (df5['专业：理工']+df5['专业：医科']+df5['专业：农科']) / (df5['专业：理工']+df5['专业：医科']+df5['专业：社科']+df5['专业：农科']+df5['专业：人文'])
print(df5[['社科比例','人文比例','理工农医比例']])
print('-'*60)
# 计算每年专业大类分布数据
```
