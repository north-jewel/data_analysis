```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
mpl.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
%matplotlib inline
```

```python
def weiji_1802c():
    a = [39,8,0,1,6]
    b = range(len(a))
    plt.figure()
    plt.title('1802c班违纪情况')
    plt.xticks(b,(['未带电脑总数','未带手机总数','未关投影次数','未关门窗次数','校查违纪']))
    plt.annotate('39',xy=(0,40))
    plt.annotate('8',xy=(1,10))
    plt.annotate('0',xy=(2,1))
    plt.annotate('1',xy=(3,1))
    plt.annotate('6',xy=(4,7))
    plt.grid(axis='y')
    ax = plt.gca()
    ax.spines['left'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    plt.bar(b,a)
    plt.show()
weiji_1802c()
```

```python
weiji_1802c()
```

```python
def contrast():
    np.random.seed(500)
    a = np.random.randint(0,300,9)
    b = range(len(a))
    c = np.random.randint(0,50,9)
    d = np.random.randint(0,50,9)
    e = np.random.randint(0,50,9)
    f = np.random.randint(0,50,9)
    plt.figure()
    plt.title('9.25--10.25月度违纪对比')
    plt.xticks(b,(['1802c','1803c','1804c','1805c','1806c1','1806c2','1807c','1808c1','1808c2']))
    plt.grid(axis='y')
    plt.xlabel(['未带电脑总数','未带手机总数','未关投影次数','未关门窗次数','校查违纪'])
    ax = plt.gca()
    ax.spines['left'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    plt.bar(b,a,label='未带电脑总数')
    plt.bar(b,c,bottom=a,color='orange',label='未带手机总数')
    plt.bar(b,d,bottom=a,color='gray',label='未关投影次数')
    plt.bar(b,e,bottom=a,color='yellow',label='未关门窗次数')
    plt.bar(b,f,bottom=a,color='red',label='校查违纪')
    plt.legend(loc='upper right')
contrast()
```
