```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
def method():
    plt.figure()
    
    plt.title('1802C班违纪情况')
    
    ax=plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    
    N=5
    values=(39,8,0,1,6)
    index=np.arange(N)
    plt.xticks(index, ('未带电脑总数', '未带手机总数', '未关投影仪总数', '未关门窗次数', '校查违纪'))
    plt.yticks(np.linspace(0,45,10))
    
    for x,y in zip(index,values):
        plt.text(x+0.2,y+0.3,'%.0f'%y,ha='center',va='bottom')
    
    plt.grid(axis='y')
    plt.bar(index,values)
    plt.show()
    
method()
```
