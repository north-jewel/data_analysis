```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

# 第一题

```python
def method():
    plt.figure()
    
    plt.title('1802C班违纪情况')
    
    ax=plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    
    N=5
    values=(39,8,0,1,6)
    index=np.arange(N)
    plt.xticks(index, ('未带电脑总数', '未带手机总数', '未关投影仪总数', '未关门窗次数', '校查违纪'))
    plt.yticks(np.linspace(0,45,10))
    
    for x,y in zip(index,values):
        plt.text(x+0.2,y+0.3,'%.0f'%y,ha='center',va='bottom')
    
    plt.grid(axis='y')
    plt.bar(index,values)
    plt.show()
    
method()
```

# 第二题

```python
plt.figure()

plt.title('9.25-10.25月度违纪对比')
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
    
N=10
one=np.random.randint(0,260,10)
two=np.random.randint(10,16,10)
three=np.random.randint(10,20,10)
four=np.random.randint(10,13,10)
five=np.random.randint(3,13,10)
ind=np.arange(N)

plt.xticks(ind, ('1802C', '1803C', '1804C', '1805C', '1806C1','1806C2','1807C1','1807C2','1808C1','1808C2'))
plt.yticks(np.linspace(0,300,7))


plt.bar(ind,one,color='b',label='未带电脑总数')
plt.bar(ind,two,color='orange',bottom=one,label='未带手机总数')
plt.bar(ind,three,color='gray',bottom=one+two,label='未关投影仪总数')
plt.bar(ind,four,color='y',bottom=one+two+three,label='未关门窗次数')
plt.bar(ind,five,color='r',bottom=one+two+three+four,label='校查违纪')


plt.legend(loc=8,ncol=5,markerscale=0.1)

plt.grid(axis='y')
plt.show()
```
