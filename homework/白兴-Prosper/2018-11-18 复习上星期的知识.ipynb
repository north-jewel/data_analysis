```python
x=np.linspace(-np.pi,np.pi,256,endpoint=True)
```

```python
c,s=np.cos(x),np.sin(x)
```

```python
plt.figure()
plt.plot(x,c,'r--',label='cos')
plt.plot(x,s,'k--',label='sin')
plt.plot(x,np.abs(x)<0.5)

ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

ax.xaxis.set_ticks_position('bottom')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$'])
ax.yaxis.set_ticks_position('left')
plt.yticks(np.linspace(-1,1,5))

for label in ax.get_yticklabels()+ax.get_xticklabels():
    label.set_fontsize(18)
    label.set_bbox(dict(facecolor='m',edgecolor='r',alpha=0.2))
ax.grid(color='y')
plt.fill_between(x,np.abs(x)<0.5,c,c>0.1,color='green')

t=1
plt.plot([t,t],[0,np.cos(t)],'y',linewidth=3,linestyle='--')
plt.annotate('y=cos(1)',xy=(t,np.cos(1)),xytext=(+10,+20),
            textcoords='offset points',arrowprops=dict(arrowstyle='->',
            connectionstyle='arc3,rad=.8'))
plt.legend()
plt.show()
```

# 散点图 scatter

```python
fig=plt.figure()

n=128
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)
T=np.arctan2(X,Y)

plt.axes([0.025,0.025,0.95,0.95],facecolor='r')
plt.scatter(X,Y,s=175,c=T,alpha=.5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.axis()
plt.title('scatter')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 柱状图

```python
fig.add_subplot(332)
n=10
X=np.arange(n)
Y1=(1-X/float(n)*np.random.uniform(0.5,1.0,n))
Y2=(1-X/float(n)*np.random.uniform(0.5,1.0,n))

plt.bar(X,Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='w')

for x,y in zip(X,Y1):
    plt.text(x+0.1,y+0.02,'%.2f'%y,ha='center',va='bottom')
for x,y in zip(X,Y2):
    plt.text(x+0.1,-y-0.02,'%.2f'%y,ha='center',va='top')
plt.show()
```

# 极坐标图

```python
from pylab import *
mpl.rcParams['font.sans-serif'] =['SimHei']

fig=plt.figure()
n=6
theta=np.arange(0,2*np.pi+0.01,2*np.pi/n)
radii=np.array([6]*7)
plt.polar(theta,radii)
plt.xticks(np.arange(6),('魔法攻击力','独立攻击力','战力','力量','智力','属性强化'))

plt.show()
```

# 绘制line

```python
x=np.linspace(-np.pi,np.pi,255,endpoint=True)
```

```python
c,s=np.cos(x),np.sin(x)
```

```python
plt.figure()

plt.plot(x,c,color='blue',linewidth=3.8,linestyle='--',label='cos',alpha=0.6)
plt.plot(x,s,'g--')

plt.title('sin&cos')
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

ax.xaxis.set_ticks_position('bottom')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
           [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$'])

ax.yaxis.set_ticks_position('left')
plt.yticks(np.linspace(-1,1,5))

for label in ax.get_yticklabels()+ax.get_xticklabels():
    label.set_fontsize(10)
    label.set_bbox(dict(facecolor='red',edgecolor='k',alpha=0.3))
ax.grid(color='y')

plt.fill_between(x,np.abs(x)<0.5,c,c>0.1,color='k')

t=1
plt.plot([1,1],[0,cos(t)],'y--',linewidth=3)
plt.annotate('y=cos(1)',xy=(t,cos(t)),xycoords='data',
             xytext=(+10,+20),textcoords=('offset points'),
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.8'))
plt.legend(loc=2)
plt.show()
```
