###  最贵的车是哪款车

```python
che = list(df['Officialprice'].values)
che
che_zg = []
for i in che:
    che_zg.append(eval(i.replace('万','').split('-')[-1]))
# print(che_zg)
che_zg_a = sorted(che_zg,reverse = True)[0]
df.iloc[che_zg.index(che_zg_a)]
```

### 哪些车车主评价比较高

```python
a = df.groupby('Ownerevaluate').count().Carname.sort_values(ascending = False)
a
youxiu = df[df['Ownerevaluate'] == '优秀']
print(list(youxiu.Carname.values))
```

### 那种车型最多

```python
che_xing = df.groupby('Rank').count().Carname
zd_che_xing = che_xing.sort_values(ascending = False)
plt.figure(figsize=(13,5),dpi=100)
ax = plt.gca()
plt.plot(zd_che_xing)
plt.title('车型数量')
plt.xlabel('车型')
plt.ylabel('数量')
plt.grid(axis='y')
plt.show()
```
