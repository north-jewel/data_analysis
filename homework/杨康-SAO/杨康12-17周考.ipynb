```python
import pandas as pd
```

```python
df = pd.read_csv(r'pcauto_info.csv')
```

```python
df.head()
```

# 大排量车有哪些？

```python
df2 = df[df.displacement.str.endswith('T'and 'L')]
df_list = list(df2.displacement.str.replace('L','').str.replace('T','').str.split(' '))
```

```python
list1 = []
for i in range(len(df_list)):
    list1.append(float(df_list[i][-1]))
df2['displacement_sort'] = list1
df2.sort_values('displacement_sort',ascending=False).head(10)
```

# 小排量车有哪些？

```python
df2 = df[df.displacement.str.endswith('T'and 'L')]
df_list = list(df2.displacement.str.replace('L','').str.replace('T','').str.split(' '))
```

```python
list1 = []
for i in range(len(df_list)):
    list1.append(float(df_list[i][-1]))
df2['displacement_sort'] = list1
df2.sort_values('displacement_sort').head(10)
```

# 最贵的是什么车？

```python
a = list(df2.officialprice.str.replace('万','').str.split('-'))
```

```python
list2 = []
for i in range(len(a)):
    list2.append(float(a[i][-1]))
#print(list1)
df2['officialprice_sort']=list2
df2.sort_values('officialprice_sort',ascending=False).head(10)
```
