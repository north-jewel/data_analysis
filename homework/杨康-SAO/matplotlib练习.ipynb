```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
x = np.linspace(-np.pi,np.pi,20,endpoint=True)
```

```python
y=np.sin(x)
```

```python
z = np.cos(x)
```

```python
plt.figure(1)
```

```python
plt.title('sin & cos') # title单独画会单独显示
plt.plot(x,y,color='green',linewidth=4.0,linestyle='--',label='sin',alpha=0.5)
plt.legend()
plt.plot(x,z,'co')
```

### [-5,5] y=x**2 黄线 点图 标题为'y=x**2' 图例为平方图

```python
x = np.linspace(-5,5,20,endpoint=True)
```

```python
y=x**2
```

```python
plt.title('y=x**2')
plt.plot(x,y,'y:',label='平方图')
plt.legend()
```
