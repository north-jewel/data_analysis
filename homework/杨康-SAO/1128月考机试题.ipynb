```python
def discipline_hist(class_dict,color_list):

    class_names = list(class_dict.keys())
    class_data = list(class_dict.values())
    #print(class_data)

    fig,ax = plt.subplots()
    plt.title('9.25~10.25月度违纪对比')
    
    discipline_list = []
    label_list = list(class_data[0].keys())
    for i in range(len(class_dict)):
        discipline_list.append(list(class_data[i].values()))
    discipline_array = np.array(discipline_list)
    print(discipline_list)
    print(discipline_array)
    cumsum_array = np.cumsum(discipline_array,axis=1)
    discipline_items = discipline_array.shape
    index_list = list(range(1,discipline_items[1]))
    ax.bar(class_names,discipline_array[:,0],label=label_list[0])
    for i,c,x in zip(index_list,color_list,label_list[1:]):
        ax.bar(class_names,discipline_array[:,i],color=c,bottom=cumsum_array[:,i-1],label=x)
    plt.legend(loc='lower center',ncol=5,bbox_to_anchor=(0.5, -0.2),fontsize=8)
```

```python
discipline_hist(discipline_class,color_list)
```
