```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
class_data = {'未带电脑总数':39,
                    '未带手机总数':8,
                    '未关投影仪次数':0,
                    '未关门窗次数':1,
                    '校查违纪':6,}
```

```python
def discipline_hist(class_data):

    group_names = list(class_data.keys())
    group_data = list(class_data.values())
    fig,ax = plt.subplots()
    plt.title('1802C班违纪情况')
    ax.bar(group_names,group_data,)
    x = np.arange(len(class_data))
    for i,y in zip(x,group_data):
        #print(i,y)
        plt.text(i,y,group_names[i]+':'+str(y),ha='center',va='bottom')
discipline_hist(class_data)
```
