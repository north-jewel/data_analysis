```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']  # 用来正常显示中文标签  
plt.rcParams['axes.unicode_minus']=False  # 用来正常显示负号
def paint(data,headline):
    #np.random.seed(100)
    fig,ax = plt.subplots()
    x = np.arange(5)
    plt.title(headline)
    plt.xticks(x)
    label = ['未带电脑总数','未带手机总数','未关投影次数','未关门窗次数','校查违纪']
    ax.set_xticklabels(label)
    ax.set_ylim([0, 50])
    ax.spines['right'].set_color('None')
    ax.spines['top'].set_color('None')
    #plt.yticks([0,10,20,30,40,50])
    plt.bar(x,data)
    for a,b,c in zip(list(x),data,label):
        plt.annotate(c+str(b), xy=(a,b), xytext=(0.4, 30),textcoords = 'offset points',arrowprops=dict(facecolor='w', shrink=0.06))
    plt.show()
paint(data = np.random.randint(20,50,5), headline = '1802C班违纪情况')
```
