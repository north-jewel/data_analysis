```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']  # 用来正常显示中文标签  
plt.rcParams['axes.unicode_minus']=False  # 用来正常显示负号
def paint(data):
    fig,ax = plt.subplots()
    x = np.arange(10)
    plt.title('9.25-10.25月度违纪对比')
    plt.xticks(x)
    ax.set_xticklabels(['1802C','1803C','1804C','1805C','1806C1','1806C2','1807C1','1807C2','1808C1','1808C2'])
    ax.set_ylim([0, 300])
    ax.spines['right'].set_color('None')
    ax.spines['top'].set_color('None')
    plt.grid(axis ='y' )
    #for a in zip(list(data['data_mobile']),list('data_computer'),list('data_projector'),list('data_doors'),list('data_Discipline'))
    plt.bar(x,data['data_mobile'],label = '没带手机')
    plt.bar(x,data['data_computer'],bottom=data['data_mobile'],label = '没带电脑')
    plt.bar(x,data['data_projector'],bottom=data['data_computer']+data['data_mobile'],label = '没关投影仪')
    plt.bar(x,data['data_doors'],bottom=data['data_projector']+data['data_computer']+data['data_mobile'],label = '没关门窗')
    plt.bar(x,data['data_Discipline'],bottom=data['data_doors']+data['data_projector']+data['data_computer']+data['data_mobile'],label = '校查违纪')
    plt.legend()
    plt.show()
paint(data = {
        'data_mobile' : np.random.randint(20,21,10),
        'data_computer' : np.random.randint(20,21,10),
        'data_projector' : np.random.randint(20,21,10),
        'data_doors' : np.random.randint(20,21,10),
        'data_Discipline' : np.random.randint(20,21,10)})
```
