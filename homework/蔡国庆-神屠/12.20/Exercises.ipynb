### Step 5. Create a histogram of the top 5 items bought

```python

# x = chipo.item_name
# letter_name = Conter(x)

# chipo.head()

chipo.drop(chipo[chipo.item_price.isnull()].index,inplace = True)
#chipo.item_price = [float(value[1:]) for value in chipo.item_price]
chipo.head()
df = chipo.loc[:,['item_name','item_price']]
df1 = df.sort_values('item_price',ascending = False).head(5)
plt.barh(df1.item_name,df1.item_price)
plt.show()
```
