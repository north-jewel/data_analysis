## 什么车价钱最贵


```python
df1 = df.loc[:,'car_name':'official_price']
price = df1.official_price.apply(lambda x:x[:-1].split('-'))
price_list = []
for i in price.values:
    try:
        price_list.append(float(i[1]))
    except IndexError:
        price_list.append(float(i[0]))
len(price_list)
df1.official_price = price_list
num = str(df1.official_price.max())
df1[df1.official_price.isin([num])]
```

## 什么车的排量最高

```python
df2 = df.loc[:,['car_name','displacement']]
df2.index = df2.car_name
df2
df_data = df2.displacement.str.replace(' ','').str.replace('T','').str.replace('L','').str.replace('--','').str.replace('纯电动','')
df_data
df2_max = df_data.str[-3:].sort_values(ascending = False)
df2_max.head(50)
```

## 那些车排量比较小

```python
df2_min = df_data.str[:3].sort_values(ascending = True)
min_list = []
for i in df2_min.values:
    if i != '':
        if float(i)< 1.5:
            min_list.append(i)
min_list
df2_min[df2_min.isin(min_list)]
```
