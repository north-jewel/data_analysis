```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = {'brain':np.random.randint(2,250,10),'mobile':np.random.randint(2,10,10),
        'shadow':np.random.randint(2, 10, 10),'sill' : np.random.randint(2,10,10),
        'rule':np.random.randint(2, 10, 10)}
all_class = ['1802c','1803c','1804c','1805c','1806c','1807c','1808c','1809c',
                     '1803a','1804a']
df = pd.DataFrame(data,index=all_class)
```

```python
fig, ax = plt.subplots()
x = np.arange(10)
plt.title('9-25-10.25月度违纪对比')
plt.ylim(0,300)
#plt.xlim(-1,10)
plt.xticks(x,all_class)
plt.bar(x,data['brain'],label='电脑')
plt.bar(x,data['mobile'],bottom=data['brain'],label='手机')
plt.bar(x,data['sill'],bottom=data['brain']+data['mobile'],label='未关窗')
plt.bar(x,data['shadow'],bottom=data['brain']+data['mobile']+data['sill'],label='投影')
plt.bar(x,data['rule'],bottom=data['brain']+data['mobile']+data['sill']+data['shadow'],label='违纪')
plt.legend(loc = 'upper left')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


data = {'未带电脑数':np.random.randint(2,250,10),'未带手机数':np.random.randint(2,10,10),
        '未关投影数':np.random.randint(2, 10, 10),'未关窗数' : np.random.randint(2,10,10),
        '违规':np.random.randint(2, 10, 10),
        '班级':['1802c','1803c','1804c','1805c','1806c','1807c','1808c','1809c','1803a','1804a']}

def jishi(data,title=None):
    fig, ax = plt.subplots()
    if title == None:
        title = '未命名'
    for i in ax.spines:
        ax.spines[i].set_color('none')
    #ax = gca()
    x = np.arange(10)
    plt.title(title)
    plt.ylim(0,300)
    plt.grid(axis='y')
    ax.set_axisbelow(True)
    plt.xticks(x,data['班级'])
    plt.bar(x,data['未带电脑数'],label='电脑')
    plt.bar(x,data['未带手机数'],bottom=data['未带电脑数'],label='手机')
    plt.bar(x,data['未关投影数'],bottom=data['未带电脑数']+data['未带手机数'],label='未关窗')
    plt.bar(x,data['未关窗数'],bottom=data['未带电脑数']+data['未带手机数']+data['未关投影数'],label='投影')
    plt.bar(x,data['违规'],bottom=data['未带电脑数']+data['未带手机数']+data['未关投影数']+data['未关窗数'],label='违纪')
    plt.legend(loc=3,ncol=15,bbox_to_anchor=(-0.01,-0.2))
    plt.show()

jishi(data,)
```
