# 哪种车的排量最大

```python
df1 = df.loc[:,('carname','displacement')]
```

```python
b = []
for i in df1.displacement:
    for x in eval(i):
        if x != '纯电动':
            if x[:-1] != '-':
                b.append(x[:-1])
c = sorted(b)[-1]#最大排量

for i in df1.displacement:
    if c in i:
        d = i
df[df.displacement==d]
```

# 那种车最贵

```python
df2 = df.loc[:,('carname','officialprice')]
```

```python
d = []
for i in df2.officialprice:
    d.append(i[:-1])
f = []
for i in d:
    if len(i.split('-'))==2:
        f.append(float(i.split('-')[1]))
    else:
        f.append(float(i.split('-')[0]))
f
max_price = sorted(f)[-1]
for i  in df2.officialprice:
    if str(max_price) in i:
        max_price = i
```

```python
max_price
df[df.officialprice==max_price]
```
