```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Shilin\Desktop\info.csv')
df
```

```python
BIG = []
SAMLL = []
for i in df.Displacement:
    BIG.append(str(i).replace('T','').replace('纯电动','').replace('L','').replace(", ''",'').replace("'",'').replace(',','')[:4])
    SAMLL.append(str(i).replace('T','').replace('纯电动','').replace('L','').replace('\r\n','').replace(", ''",'').replace("'",'')[-4:])
```

```python
df
```

```python
df['BIG'] = BIG
df['SAMLL'] = SAMLL
```

# 大排量车

```python
s1 = df.sort_values('BIG',ascending=False).head()
s1
```
