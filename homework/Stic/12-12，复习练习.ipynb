```python
import matplotlib.pyplot as plt
def mandelbrot(h,w,maxit =20):
    """hw"""
    y,x = np.ogrid[-1.4:1.4:h*1j,-2:0.8:w*1j]
#     print(y,x)
    c = x+y*1j
    print(c)
    z = c
    divtime = maxit + np.zeros(z.shape,dtype = int)
    for i in range(maxit):
        z = z**2 + c
        diverge = z*np.conj(z) > 2**2   #
        div_now = diverge & (divtime == maxit)  #
        divtime[div_now] = i                  #
        z[diverge] = 2                        #
        
    return divtime

plt.imshow(mandelbrot(400,400))
plt.show()
```

```python
def checkPossibility(nums):
    """
    :type nums: List[int]
    :rtype: bool
    """
    nums_list = [1]
    sums = len(nums)
    if sums > 1:
        for i in range(len(nums))[1:]:
            if i+1 == sums:
                break
            else:
                a = nums[i-1]
                b = nums[i]
                c = nums[i+1]
                if a > b and a > c and b > c:
                    return False
                if a > c and b > c:
                    nums_list.append(1)
                if a > b or b > c:
                    nums_list.append(1)
                    nums[i] = c
                if sum(nums_list) == 3:
                    return False
    if sum(nums_list) == 3:
        return False
    else:
        return True
```

```python
def checkPossibility(nums):
    """
    :type nums: List[int]
    :rtype: bool
    """
    b = 0
    if len(nums) > 2:
        for i in range(len(nums))[1:-1]:
            n = nums[:i+1]
            if  n == sorted(n):
                continue
            else:
                if b == 1:
                    return False
                b = b + 1
                if nums[i-1] >= nums[i+1]:
                    nums[i-1] = nums[i-2]
                else:
                    del nums[:i-1]
        return True
    else:
        return True
```

```python
print(checkPossibility([3,4,2,3]))    #F
print(checkPossibility([1,2,4,5,3]))  #T
print(checkPossibility([4,2,3]))      #T
print(checkPossibility([4,2,1]))      #F
print(checkPossibility([-1,4,2,3]))   #T
print(checkPossibility([2,3,3,2,4]))  #T
```
