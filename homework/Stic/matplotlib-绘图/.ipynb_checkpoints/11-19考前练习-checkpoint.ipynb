```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
labels='a','b','c','d'
sizes=[20,15,35,30]
explode=(0,0,0,0.1)
fig = plt.figure()
ax = fig.add_subplot(1,2,1)
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.0f%%',shadow=True,startangle=180)
ax.axis('equal')
ax = fig.add_subplot(1,2,2)
plt.pie(sizes,explode=explode,labels=labels,autopct='%.0f%%',shadow=False,startangle=180)
ax.axis('equal')
plt.show()
```

```python
n = 20
z=np.ones(n)
z[-1]*=2
# plt.figure(10)
fig,ax=plt.subplots(2,2)
ax[0,0].pie(z,explode=z*0.05,colors=['%f'%(i/float(n)) for i in range(n)],    #白黑色
        labels=['%.2f'%(i/float(n))for i in range(n)])
ax[0,0].set_aspect('equal')
ax[0,1].pie(z,explode=z*0.05,    #默认彩色
        labels=['%.2f'%(i/float(n))for i in range(n)])
ax[0,1].set_aspect('equal')
ax[1,0].pie(z,explode=z*0.05,colors=('b','g','r','snow','aqua'),    #指定颜色
        labels=['%.2f'%(i/float(n))for i in range(n)],radius=2)   #radius 图的大小，默认1
ax[1,0].set_aspect('equal')
ax[1,1].pie(z, explode=None, labels=['%.2f'%(i/float(n))for i in range(n)],  
    colors=('b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'),  
    center = (0, 0.1), frame = True ) #center(x,y) 圆中心的位置
ax[1,1].set_aspect('equal')
# plt.show()
```
