```python
an = np.linspace(0, 2 * np.pi, 100)
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[0, 0].set_title('not equal, looks like ellipse', fontsize=10)

axs[0, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[0, 1].axis('equal')
axs[0, 1].set_title('equal, looks like circle', fontsize=10)

axs[1, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[1, 0].axis('equal')
axs[1, 0].axis([-3, 3, -3, 3])
axs[1, 0].set_title('still a circle, even after changing limits', fontsize=10)

axs[1, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[1, 1].set_aspect('equal', 'box')


axs[1, 1].set_title('still a circle, auto-adjusted data limits', fontsize=10)

fig.tight_layout()

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-np.pi,np.pi,266,endpoint=True)
a=np.sin(x)
b=np.cos(x)
```

```python
plt.figure()  #所有图的容器
plt.plot(x,a,linewidth=3.0,linestyle='--',color='gray',label='sin',alpha=0.65)
plt.plot(x,b,linewidth=3.0,linestyle='--',color='red',label='cos',alpha=0.65)
plt.title('sin & cos')   #标题
# plt.subplots(2, 2)
ax = plt.gca()
ax.spines['bottom']. set_position(('data',0))   #改变线框，移动下面的线到0原点
ax.spines['bottom'].set_color('red')
ax.spines['left']. set_position(('data',0))
ax.spines['left'].set_color('gray')
ax.spines['top']. set_color('none')
ax.spines['right']. set_color('none')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
         [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$\pi$'])
plt.yticks([1,0.5,0,-0.5,-1])    #改变y轴的点
plt.tight_layout()  #紧密布局
plt.legend()   #图列
# plt.show()
for l in ax.get_xticklabels()+ax.get_yticklabels():
    l.set_fontsize(18)     #字体大小
    l.set_bbox(dict(edgecolor='white',alpha=0.65))   #字体边框的颜色，透明度
    l.set_color('red')     #字体颜色
    l.set_backgroundcolor('pink')   #字体背景
ax.grid()  #网格线
ax.fill_between(x,b*(np.abs(x)<0.5),color='green',alpha=0.25)
ax.fill_between(x,0,a*(x<np.pi/4)-a*(x<0)+b*(x>np.pi/4)-b*(x>np.pi/2),color='blue',alpha=0.5)
ax.fill_between(x,a,color='gray',alpha=0.25)
ax.fill_between(x,x>0,b,b<0.5,color='red',alpha=0.25)
# ax.fill_between(x,0,b,b<0.5,color='red',alpha=0.25)
ax.fill_between(x,np.abs(x)<0.5,b,abs(x-0.25)<0.25,color='black')
```

```python
def f(t):
    si=np.cos(2*np.pi*t)
    ei=np.exp(-t)
    return si*ei
```

```python
t1 = np.arange(0.0,5.0,0.2)
l = plt.plot(t1,f(t1),'o')
xmin, xmax = xlim = t1-0.25, t1-0.25
ymin, ymax = ylim = f(t1)-0.25, f(t1)+0.25

ax = plt.subplots()
ax.set(xlim=xlim, ylim=ylim, autoscale_on=False)

X = [[0,0], [1,1]]
ax.imshow(X, interpolation='bicubic', cmap=plt.cm.Blue,
          extent=(xmin, xmax, ymin, ymax), alpha=1)
plt.setp(l,markersize=30)
# plt.setp(l,markerfacecolor='C0')
# plt.minorticks_on()
# plt.colormaps('gist_earth')
ax.set_aspect('auto')
# plt.imshow()
# plt.show()
```

```python
np.random.seed(19680801)


def gbar(ax, x, y, width=0.5, bottom=0):
    X = [[.6, .6], [.7, .7]]
    for left, top in zip(x, y):
        right = left + width
        ax.imshow(X, interpolation='bicubic', cmap=plt.cm.Blues,
                  extent=(left, right, bottom, top), alpha=1)


xmin, xmax = xlim = 0, 10
ymin, ymax = ylim = 0, 1

fig, ax = plt.subplots()
ax.set(xlim=xlim, ylim=ylim, autoscale_on=False)

X = [[.6, .6], [.7, .7]]
ax.imshow(X, interpolation='bicubic', cmap=plt.cm.copper,
          extent=(xmin, xmax, ymin, ymax), alpha=1)

N = 10
x = np.arange(N) + 0.25
y = np.random.rand(N)
gbar(ax, x, y, width=0.7)
ax.set_aspect('auto')
plt.show()
```

```python
plt.figure()
# fig.add_subplot()
n=10
X=np.arange(n)
Y1 = (1-X/float(n)*np.random.uniform(0.5,1.0,n))
Y2 = (1-X/float(n)*np.random.uniform(0.5,1.0,n))
plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')
for x,y in zip(X,Y1):
    plt.text(x+0.1,y,'%.2f'%y,ha='center',va='bottom') # 给上边添加注释
for x,y in zip(X,Y2):
    plt.text(x,-y-0.05,'%.2f'%y,ha='center',va='top') # 给下边添加注释
plt.tight_layout()
plt.show()
```
