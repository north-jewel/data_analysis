```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
def rule(name,arg1,arg2,arg3,arg4,arg5):
    '''
    name:班级名字
    arg:违纪次数
    '''
    a = [arg1,arg2,arg3,arg4,arg5]
    name = '{}班违纪情况'.format(name)
    plt.figure(figsize=(8,6))
    plt.bar([1,2,3,4,5],a)
    plt.annotate('未带电脑总数，{}'.format(arg1),xy=(1,arg1*(2/3)),xytext=(0.5,arg1+0.5),arrowprops=dict(facecolor='white',shrink=0.05))
    plt.annotate('未带手机总数，{}'.format(arg2),xy=(2,arg2*(2/3)),xytext=(1.5,arg2+0.5),arrowprops=dict(facecolor='white',shrink=0.05))
    plt.annotate('未关投影仪次数，{}'.format(arg3),xy=(3,arg3*(2/3)),xytext=(2.5,arg3+0.5),arrowprops=dict(facecolor='white',shrink=0.05))
    plt.annotate('未关门窗次数，{}'.format(arg4),xy=(4,arg4*(2/3)),xytext=(3.5,arg4+0.5),arrowprops=dict(facecolor='white',shrink=0.05))
    plt.annotate('校查违纪，{}'.format(arg5),xy=(5,arg5*(2/3)),xytext=(4.5,arg5+0.5),arrowprops=dict(facecolor='white',shrink=0.05))
    ax = plt.gca()
    ax.spines['left'].set_color('white')
    ax.spines['top']. set_color('none')
    ax.spines['right']. set_color('none')
    plt.title(name)
    plt.xticks([1,2,3,4,5],['未带电脑总数','未带手机总数','未关投影仪次数','未关门窗次数','校查违纪'])
```

# 月考机试修改-方法优化

```python
def rule(name,*a,list_r = ['未带电脑总数','未带手机总数','未关投影仪次数','未关门窗次数','校查违纪']):
    '''
    name:班级名字
    arg:违纪次数
    '''
    al = len(a)
    name = '{}班违纪情况'.format(name)
    plt.figure(figsize=(8,6))
    plt.bar(np.arange(al),a)
    for i in range(al):
        plt.annotate('{}，{}'.format(list_r[i],a[i]),xy=(i,a[i]*(2/3)),xytext=(i-0.5,a[i]+0.5),arrowprops=dict(facecolor='white',shrink=0.05))
    ax = plt.gca()
    ax.spines['left'].set_color('white')
    ax.spines['top']. set_color('none')
    ax.spines['right']. set_color('none')
    plt.title(name)
    plt.xticks(np.arange(al),list_r)
```

```python
rule('1802c',20,8,0,1,6)
```

```python
def rule2(name,list_class=['1802C','1803C','1804C','1805C','1806C1','1806C2','1807C1','1807C2','1808C1','1808C2','1809C'],
          list_r = ['校查违纪', '未带电脑总数','未关投影仪次数','未关门窗次数','未带手机总数'],
          *a):
    ale = len(a)
#     print(a)
    list_color = ['red','blue','green','black','gray','pink','yellow','orange','cyan','purple']
#     list_color = np.random.randint(0,256,100)
    list_c = []
    for i in range(ale):
#         print(i)
        if i == 0:
            plt.bar(range(1, 12), a[i], color=list_color[i])
            list_c.append(a[i])
        else:
            plt.bar(range(1, 12), a[i],bottom=list_c[i-1],color=list_color[i])
            c = list_c[i-1]+a[i]
            list_c.append(c)
#     print(list_c)
    plt.title('{}月度对比'.format(name))
    plt.xticks(range(1, 12),list_class)
    plt.legend(list_r,loc=8,ncol=5,bbox_to_anchor=(0.5,-0.2))  #loc:
```

```python
a1 = np.array([50,100,20,30,37,16,165,47,89,45,146])
a2 = np.array([20,110,25,35,15,15,15,15,56,56,78])
a3 = np.array([60,120,35,45,45,15,15,43,38,56,70])
a4 = np.array([80,130,45,65,25,15,78,15,56,56,15])
a5 = np.array([100,150,50,30,85,15,38,15,56,56,35])
rule2('9.25-10.25',['1802C','1803C','1804C','1805C','1806C1','1806C2','1807C1','1807C2','1808C1','1808C2','1809C'],['校查违纪', '未带电脑总数','未关投影仪次数','未关门窗次数','未带手机总数'],a1,a2,a3,a4,a5)
```
