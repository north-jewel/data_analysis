```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from pylab import mpl 
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
def bar(x,y,title):
    fig=plt.figure(figsize=(9, 6.5))
    ax=fig.add_subplot(111)
    a = []
    for i in range(len(x)*2+1):
        if i%2 == 0:
            a.append(i)
    plt.xticks(a[1:])
    ax.set_xticklabels(x)
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.spines['bottom'].set_color('none')
    plt.grid(axis = 'y')
    ax.set_axisbelow(True)
    ax.set_title(title,fontdict = {'fontsize':20})
    for i in range(len(y)):
        plt.bar(i*2+2,y[i][0],color = 'blue', align = 'center',width = 1)
        plt.bar(i*2+2,y[i][1],bottom = y[i][0],color = 'orange', align = 'center',width = 1)
        plt.bar(i*2+2,y[i][2],bottom = y[i][0]+y[i][1],color = 'gray', align = 'center',width = 1)
        plt.bar(i*2+2,y[i][3],bottom = y[i][0]+y[i][1]+y[i][2],color = 'y', align = 'center',width = 1)
        plt.bar(i*2+2,y[i][4],bottom = y[i][0]+y[i][1]+y[i][2]+y[i][3],color = 'r', align = 'center',width = 1)

    fig1, ax1 = plt.subplots() 
    plt.bar(30,0,label = '未带电脑总数',color = 'blue')
    plt.bar(30,0,label = '未带手机总数',color = 'orange')
    plt.bar(30,0,label = '未关投影仪次数',color = 'gray')
    plt.bar(30,0,label = '未关门窗次数',color = 'y')
    plt.bar(30,0,label = '校查违纪',color = 'r')
    plt.legend(loc = 'upper center',ncol = 5)
    ax1.spines['right'].set_color('none')
    ax1.spines['top'].set_color('none')
    ax1.spines['left'].set_color('none')
    ax1.spines['bottom'].set_color('none')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    
    
x = ['1802c','1803c','1804c','1805c','1806c1','1806c2','1807c1','1807c2','1808c1','1808c2']
y = np.random.randint(0,50,50).reshape(10,5)
title = '评分'
bar(x,y,title)
```
