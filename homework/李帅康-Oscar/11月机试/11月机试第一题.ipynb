```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from pylab import mpl 
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
def bar(x_name,y_size,data,title):
    '''
    'x_name':x轴上的名称列表，
    'y_size':沿轴刻度的最大值,
    'data':x轴上每个的高度,
    'title':标题名称
    '''
    fig, ax = plt.subplots()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.spines['bottom'].set_color('none')
    plt.grid(axis = 'y')
    ax.set_axisbelow(True)
    ax.set_title(title,fontdict = {'fontsize':20}) 
    ax.set_ylim([0, y_size])
    ax.set_xticks(np.arange(1,len(x_name)*2))
    for i in range(len(x_name)):
        plt.bar(i+1,data[i],color = 'b')
        ax.set_xticklabels(x_name)
    for a,b in zip(range(1,len(x_name)+1),data):
        plt.text(a, b+3, b, ha='center', va= 'bottom',fontsize=15)

x_name = ['未带电脑总数', '未带手机总数', '未关投影仪次数','未关门窗次数','校查违纪']    
y_size = 100
data = [40,31,60,11,22]
title = '1802C违纪情况'
bar(x_name,y_size,data,title) 
```
