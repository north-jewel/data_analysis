#### 将官方价转换为数字方便比较

```python
for i in df.index:
    if len(df.loc[i,'Officialprice'].replace('万','').split('-')) == 2:
        df.loc[i,'Officialprice'] = (float(df.loc[i,'Officialprice'].replace('万','').split('-')[0])+float(df.loc[i,'Officialprice'].replace('万','').split('-')[1]))/2
    else:
        df.loc[i,'Officialprice'] = float(df.loc[i,'Officialprice'].replace('万',''))
```

### 最贵的车是？

```python
top = df.groupby('Officialprice').count().tail(1).index[0]
df[df['Officialprice'] == top]
```

#### 将脱离群题的数据删除

```python
df = df.drop(df[df['Officialprice']>500].index)
```

## 各类型车什么车最贵？


```python

top_price = []
for i in list(df['Rank'].value_counts().index):
    df_dan = df[df['Rank'].str.contains(i)]
    top_price_dan = df_dan.groupby('Officialprice').count().tail(1).index[0]
    top_price.append(top_price_dan)
fig=plt.figure(figsize=(18, 6.5))
plt.bar(list(df['Rank'].value_counts().index),top_price)
```
