# 最贵的是哪款车? 

```python
df1 = df
df1.官方价 = df1.官方价.apply(lambda x:x[0:-2])
df1.官方价 = df1.官方价.apply(lambda x:x.split('-'))
df1.官方价 = df1.官方价.apply(lambda x:float(max([float(i) for i in x])))
```

```python
df1.sort_values(by = '官方价',ascending = False).head(1)
```
