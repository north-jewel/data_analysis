```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import * 
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
def school_info(info):
    x = np.arange(len(info))
    np.random.seed(1000)
    num1 = np.random.randint(1,50,10)
    num2 = np.random.randint(1,4,10)
    num3 = np.random.randint(1,3,10)
    num4 = np.random.randint(1,5,10)
    num5 = np.random.randint(1,10,10)
    fig, ax = plt.subplots()
    plt.bar(x,num1,label = '未带电脑总数')
    plt.bar(x,num2,bottom=num1,label = '未带手机总数')
    plt.bar(x,num3,bottom=num2+num1,label = '未关投影仪总数')
    plt.bar(x,num4,bottom=num3+num2+num1,label = '未关门窗次数')
    plt.bar(x,num5,bottom=num4+num3+num2+num1,label = '校查违纪')
    plt.xticks(x,info)
    plt.title('9.25-10.25月度违纪对比')
    ax.legend(loc = 'best')
    plt.show()
list1 = ['1802c','1803c','1804c','1805c','1806c1','1806c2','1807c1','1807c2','1808c1','1808c2']
school_info(list1)
```
