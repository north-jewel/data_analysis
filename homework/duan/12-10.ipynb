### 薪资分布

```python
df1 = df.dropna(how = 'any')
```

```python
df1
```

```python
sa = df1.salary
```

```python
sa
```

```python
d = []
sal = []
for i in sa:    
    c = i.replace('万/月','').replace('万/年','').replace('元/天','').replace('千/月','').replace('千以下/月','').split('-')
    #print(c)
    d.append(c)
for i in d:
    if len(i)==1:
         sal.append(i[0])
        
    else:

        s = (float(i[0])+float(i[1]))/2
        sal.append(s)
print(sal)
```

```python
df1.salary = sal
```

```python
df1
```

```python
dff = df1[df1.company_site.str.startswith('北京')]
```

```python
d = dff.groupby('company_site').count()
```

```python
d
```

```python
plt.figure()

plt.pie(d.salary,labels = d.index)
plt.legend(loc = 'best',bbox_to_anchor=(1.2,1))
```

```python
plt.figure()
for i in list(d.index):
    plt.bar(np.arange(1,15),d.salary,width = 0.5,label = i)
plt.legend(bbox_to_anchor=(1,1))
```
