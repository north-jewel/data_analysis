```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import * 
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
def class_info(a,num,info,class_name):
    
    x = np.arange(a)
    fig, ax = plt.subplots()
    plt.bar(x,num)
    plt.xticks(x,info)
    plt.title('%s班违纪情况'%(class_name))
    info_num = list(zip(info,num))
    for i in range(len(info_num)):

        plt.annotate('%s'%(str(info_num[i][0])+str(info_num[i][1])),xy = (i,int(num[i])),xycoords='data',xytext=(-30,+30),textcoords='offset points',
                            arrowprops=dict(arrowstyle='<-'))

    #plt.grid()
    plt.show()
    
class_info(5,[39,8,0,1,6],['未带电脑总数','未带手机总数','未关投影仪总数','未关门窗总数','校查违纪'],'1802c')    
```
