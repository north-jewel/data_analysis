```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
np.random.seed(100)
fig, ax = plt.subplots()
x= np.arange(1,6)
y = np.random.randint(0,45,5)

def weiji(x,y):
    plt.bar(x,y)
    plt.title('1802C班违纪情况')
    plt.grid(axis= 'y')
    
    plt.annotate('未带电脑数,17',
                 (1,10),
                 xytext =(0.6,15),arrowprops=dict(facecolor='black', shrink=0.03))
    plt.annotate('未带手机总数,24',
                 (2,25),
                 xytext =(1.7,30),arrowprops=dict(facecolor='black', shrink=0.03))
    plt.annotate('未关投影次数,3',
                 (3,5),
                 xytext =(2.7,10),arrowprops=dict(facecolor='black', shrink=0.03))
    plt.annotate('未关窗门次数,38',
                 (4,40),
                 xytext =(3.7,45),arrowprops=dict(facecolor='black', shrink=0.03))
    plt.annotate('未关窗门次数,23',
                 (5,25),
                 xytext =(4.7,30),arrowprops=dict(facecolor='black', shrink=0.03))
     
    
    ax.set_xticklabels([x,'未带电脑数','未带手机总数','未关投影次数','未关窗门次数','校查违纪'])
    ais = plt.gca()
    ais.spines['right'].set_color('none')
    ais.spines['left'].set_color('none')
    ais.spines['top'].set_color('none')
    ais.spines['bottom'].set_color('none')
    plt.show()
weiji(x,y)
```
