## 哪个牌子的车价格较高

```python
car1 = pd_car.copy()
```

```python
del car1['price_']
```

```python
car1.head(1)
```

```python
new_price = car1.Officialprice.str.replace('万','')
price_2 = new_price.str.split('-')
price_c =[]
for i in price_2:
    if len(i)==1:
        price_c.append(float(i[0]))
    else:
        price_c.append(float(i[1]))

```

```python
car1['price_c'] = price_c
```

```python
a = car1[['Brand','price_c']]
# a.Brand.value_counts()
b = a.groupby(by = ['Brand']).mean()
b.sort_values(by= 'price_c',ascending=False)
```
