### 市面上那种车型最多？？

```python
s_rank = df.groupby('Rank').count().Carname
s_rank_sort = s_rank.sort_values(ascending=False)
s_rank_sort
```

```python
plt.figure(figsize=(13,5),dpi=100)
ax = plt.gca()
plt.plot(s_rank_sort)
plt.title('车型数量')
plt.xlabel('车型')
plt.ylabel('数量')
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
plt.grid(axis='y')
plt.show()
```

### 最贵的是哪款车?

```python
jg_list = list(df['Officialprice'].values)
jg_list
```

```python
zg_jg_list = []
for jg in jg_list:
    zg_jg_list.append(eval(jg.replace('万','').split('-')[-1]))
zg_jg = sorted(zg_jg_list,reverse=True)[0]
ind = zg_jg_list.index(zg_jg)
df.iloc[ind]
```

```python
print('最贵的是{}车'.format(df.iloc[ind].Brand))
```

### 大排量车有哪些?  小排量车有哪些?

```python
s_pl = df.Displacement
s_pl
```

```python
s_pl_list = list(s_pl)
s_dpl_ind = []
s_xpl_ind = []
for i in s_pl_list:
    for ii in i.split(','):
        ii = eval(ii)
        if 'T' in ii:
            pl = eval(ii.replace('T',''))
            if pl > 1.5:
                s_dpl_ind.append(s_pl_list.index(i))
            else:
                s_xpl_ind.append(s_pl_list.index(i))
        elif 'L' in ii:
            pl = eval(ii.replace('L',''))
            if pl > 2.5:
                s_dpl_ind.append(s_pl_list.index(i))
            else:
                s_xpl_ind.append(s_pl_list.index(i))
s_dpl_ind = list(set(s_dpl_ind))
s_xpl_ind = list(set(s_xpl_ind))
print(len(s_dpl_ind))
print(len(s_xpl_ind))
```

```python
dapailiang = []
xiaopailiang = []
for i in s_dpl_ind:
    dapailiang.append(df.iloc[i].Carname)
for ii in s_xpl_ind:
    xiaopailiang.append(df.iloc[ii].Carname)
print('大排量的车有：',dapailiang)
print('小排量的车有：',xiaopailiang)
```
