```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
ytick = np.arange(0,50,5)
xtick = ['未带电脑总数','未带手机总数',
         '未关投影仪次数','未关门窗次数','校查违纪']
data = [39,8,0,1,6]
width = 0.5


ax = plt.gca()
ax.set_xticks(np.array([0,1,2,3,4,5]))
ax.set_xticklabels(xtick)
plt.yticks(ytick)

plt.xlabel('违纪行为')
plt.ylabel('违纪次数')
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

plt.bar(xtick[0],data[0],width)
ax.text(0,40,'未带电脑总数39')
plt.bar(xtick[1],data[1],width)
ax.text(1,10,'未带手机总数8')
plt.bar(xtick[2],data[2],width)
ax.text(1.5,1,'未投影仪次数8')
plt.bar(xtick[3],data[3],width)
ax.text(2.6,3,'未关门窗次数8')
plt.bar(xtick[4],data[4],width)
ax.text(3.7,7.8,'查校违纪8')

plt.grid(axis='y')
plt.title('1802C班违纪情况')
plt.show()
```

```python
def bar_1802c(x,y,data,name):
    width = 0.5
    ax = plt.gca()
    lens = len(x)
    
    plt.xlabel('违纪行为')
    plt.ylabel('违纪次数')
    
    ax.spines['left'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')

    ax.set_xticks(np.arange(lens))
    ax.set_xticklabels(x)
    ax.set_yticklabels(y)
    
    for i in range(lens):
        text = xtick[i]+str(data[i])
        xx = i-0.4
        yy = data[i]+2
        plt.bar(xtick[i],data[i],width)
        ax.text(xx,yy,text)
    
    plt.grid(axis='y')
    plt.title(name)
    return plt.show()

ytick = np.arange(0,50,5)
xtick = ['未带电脑总数','未带手机总数',
         '未关投影仪次数','未关门窗次数','校查违纪']
data = [39,8,0,3,6]
name = '1802C班违纪情况'
bar_1802c(xtick,ytick,data,name)
```
