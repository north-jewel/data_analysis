```python
aa
```

```python
pl=df.排量
```

```python
q = [1,2,3,4,5,6]
```

```python
max_pl=[]
for  i in pl:
    w=i.replace('纯电动','0.9L').split(' ')#先在w是个列表
    new_pl=[]
    for x in w:#把w变成float的列表
        if x =='--':
            pass
        else:
            new_pl.append(float(x.replace('T','').replace('L','')))
    numbers = 0
    for y in new_pl:
        if y>numbers:
            numbers = y
    max_pl.append(numbers)
```

```python
max_pl
```

```python
max_len=len(max_pl)
```

```python
max_len_i = []
for i in range(max_len):
    max_len_i.append(i)
```

```python
df.shape
```

```python
for j in range(max_len-1):
    count = 0
    for i in range(0,max_len-1-j):
        
        if max_pl[i]>max_pl[i+1]:
            max_pl[i],max_pl[i+1] = max_pl[i+1],max_pl[i]
            max_len_i[i],max_len_i[i+1] = max_len_i[i+1],max_len_i[i]
            count+=1
    if count == 0:
        break
        
    
```
