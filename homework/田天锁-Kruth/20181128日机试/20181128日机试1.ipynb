```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
mpl.rcParams['font.sans-serif'] =['SimHei']
```

```python
def plots(nd,sj,ty,mc,xc):
    x=['未带电脑总数','未带手机总数','未关投影次数','未关门窗次数','校查违纪']
    yerrs = [2,2,2,2,2]
    y=[nd,sj,ty,mc,xc]
    plt.bar(x,y,yerr =yerrs )
    plt.title('1802c班违纪情况')
    for q,w in zip(x,y):
        plt.text(q,w,q+'%.2i'%w,ha = 'center',va = 'bottom')
    plt.show()
```

```python
plots(10,20,60,70,50)
```
