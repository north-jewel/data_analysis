```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 饼图
labels = 'Cats','Birds','Dogs','Pandas'
sizes = [25,35,30,10]
explode = (0,0,0.05,0.08)

fig1,ax1 = plt.subplots()
plt.pie(sizes,explode = explode,labels = labels,
       autopct = '%1.2f%%',shadow = False,startangle = 90)
ax1.axis('equal')
plt.show()
```

```python
fig2,ax2 = plt.subplots(2,1)
n = 20
Z = np.ones(n)
Z[-1] *= 2
colors = np.array(['r','g','b','orange'])

plt.axis('equal')
ax2[0].pie(Z,explode = Z*0.01,colors = [np.random.choice(colors) for i in range(n)],
            labels = ['%.2f'%(i/float(n)) for i in range(n)])
#colors = [np.random.choice(colors) for i in range(n)]
#print(colors)
ax2[1].pie(Z,explode = Z*0.01,colors = ('r','g','b','orange'),
            labels = ['%.2f'%(i/float(n))for i in range(n)])

plt.axis('equal')
```
