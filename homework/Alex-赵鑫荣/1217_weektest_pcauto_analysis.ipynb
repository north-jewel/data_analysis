```python
#首先判断排量这一列是否有空值
s = df['displacement'].isnull()
s[s.values == True].index
```

```python
#处理汽车排量这一列，取出排量的数值，如果是区间：就取中间数、
regex = '\d+'
displace_max = []
displace_min = []
for i in df['displacement']:
#     for x in eval(i):
    i = eval(i)
    i.sort()
    if len(i) == 1:
        if i[0] == '纯电动':
            displace_max.append(0)
            displace_min.append(0)
        else:
            displace_max.append(i[0].replace('L','').replace('T',''))
            displace_min.append(i[0].replace('L','').replace('T',''))
    else:
        if i[-1] == '纯电动':
            displace_max.append(i[-2].replace('L','').replace('T',''))
            displace_min.append(0)
        else:
            displace_max.append(i[-1].replace('L','').replace('T',''))
            displace_min.append(i[0].replace('L','').replace('T',''))

df['displace_max']=displace_max
df['displace_min']=displace_min
```

```python
regex = '[0-9]*\.?[0-9]+'
df['dis_regex'] = df['displacement'].str.findall(regex)
df['d_max'] = df['dis_regex'].apply(lambda x:float(max(x)) if len(x)>0 else 0)
df['d_min'] = df['dis_regex'].apply(lambda x:float(min(x)) if len(x)>0 else 0)
df.drop('dis_regex',axis=1,inplace =True)
```

```python
#找出最贵的汽车是哪辆？
df.head()
average_list = []
for i in df['officialprice']:
    i_list = i.replace('万','').split('-')
    float_i_list = [float(x)for x in i_list]
    average_num = float_i_list[-1]
    average_list.append(average_num)
df['average_price'] = average_list
```
