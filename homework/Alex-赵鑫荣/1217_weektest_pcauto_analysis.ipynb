```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
location = r'E:\python4\jdpro\pcauto.csv'
```

```python
df = pd.read_csv(location)
```

```python
df.head(10)
```

```python
#首先判断排量这一列是否有空值
s = df['displacement'].isnull()
s[s.values == True].index
```

```python
#处理汽车排量这一列，取出排量的数值，如果是区间：就取中间数、
regex = '\d+'
displace_list = []
for i in df['displacement']:
#     for x in eval(i):
    i = eval(i)
    i.sort()
    if len(i) == 1:
        if i[0] == '纯电动':
            displace_list.append(0)
        else:
            displace_list.append(i[0].replace('L','').replace('T',''))
    else:
        if i[-1] == '纯电动':
            displace_list.append(i[-2].replace('L','').replace('T',''))
        else:
            displace_list.append(i[-1].replace('L','').replace('T',''))

df['displace_max']=displace_list
```

```python
df.head()
```

```python
#判断出排量最大的汽车
df[df['displace_max'].max() == df['displace_max']]
```

```python
#找出最贵的汽车是哪辆？
df.head()
average_list = []
for i in df['officialprice']:
    i_list = i.replace('万','').split('-')
    float_i_list = [float(x)for x in i_list]
    average_num = sum(float_i_list)/2
    average_list.append(average_num)
df['average_price'] = average_list
```

```python
#求出最贵的汽车
df[df['average_price'].max() == df['average_price']]
```

```python
#如果汽车的价钱够高、排量就一定很高么？
df.groupby(['displace_max']).apply(lambda x:x.sort_values('average_price',ascending = True))
```
