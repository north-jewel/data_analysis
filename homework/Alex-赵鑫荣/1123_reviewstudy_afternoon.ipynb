```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#饼图
labels = 'Cats','Birds','Dogs','Pandas'
sizes = [25,35,30,10]
explode = (0,0,0.05,0.08)

fig1,ax1 = plt.subplots()
print(type(fig1))
print(type(ax1))
plt.pie(sizes,explode = explode,labels = labels,
       autopct = '%1.2f%%',shadow = False,startangle=90)
ax1.axis('equal')
```

```python
#饼图
vals = [1,2,3,4]
fig,ax = plt.subplots()
#fig 返回<class 'matplotlib.figure.Figure'>对象
#ax  返回<class 'matplotlib.axes._subplots.AxesSubplot'>对象
print(type(fig))
print(type(ax))
labels = 'A','B','C','D'
colors = ['yellowgreen','gold','lightskyblue','lightcoral']
explode = (0,0.1,0,0)
ax.pie(vals,explode = explode,labels = labels,colors = colors,
      autopct = '%1.1f%%',shadow = True,startangle = 90,radius = 1.2)
ax.set(aspect = 'equal',title = '一个饼图的示例')
plt.rcParams['font.sans-serif'] = 'SimHei'
```

```python
fig2,ax2 = plt.subplots(2,2)
n = 20
Z = np.ones(n)
Z[-1] *= 2
colors = np.array(['r','g','b','orange'])

ax2[0,1].pie(Z,explode = Z*0.01,colors = [np.random.choice(colors) for i in range(n)],
             labels = ['%.2f'%(i/float(n))for i in range(n)])

ax2[1,1].pie(Z,explode = Z*0.01,colors = ('r','g','b','orange'),
            labels = ['%.2f'%(i/float(n))for i in range(n)])

ax2[1,1].set(aspect = 1)
#plt.axis('equal')
```

```python
#3D图
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.plot([1,1,1],[1,2,2],[2,2,3])
ax.plot([1,1],[1,2],[2,3])

ax.plot([2,1],[1,1],[2,2])
ax.plot([2,1],[1,2],[2,3])
ax.plot([2,1],[1,2],[2,2])
```

```python
np.random.seed(19680801)
def randrange(n,vmin,vmax):
    return (vmax-vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
n = 100

for c,m,zlow,zhigh in [('r','o',-50,-25),('b','^',-30,-5)]:
    xs = randrange(n,23,32)
    ys = randrange(n,0,100)
    zs = randrange(n,zlow,zhigh)
    ax.plot(xs,ys,zs,c = c,marker = m)
    
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_zlabel('Z label')
```

```python
#3d 散点图
from mpl_toolkits.mplot3d import Axes3D
data = np.random.randint(0,255,size = [40,40,40])
x,y,z = data[0],data[1],data[2]
ax = plt.subplot(111,projection = '3d')

ax.scatter(x[:10],y[:10],z[:10],c='y')
ax.scatter(x[10:20],y[10:20],z[10:20],c='r')
ax.scatter(x[30:40],y[30:40],z[30:40],c='g')

ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_zlabel('Z label')
```

```python
fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)
x,y = np.meshgrid(x,y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

ax.plot_surface(x,y,z,rstride = 1,cstride = 1,cmap = 'rainbow')
```
