```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\NINGMER\Desktop\info.csv')
df
```

```python
b = []
d = []
for i in df.Displacement:
    b.append(str(i).replace('T','').replace('纯电动','').replace('L','').replace(", ''",'').replace("'",'').replace(',','')[:4])
    d.append(str(i).replace('T','').replace('纯电动','').replace('L','').replace(", ''",'').replace("'",'').replace(',','')[-4:])
    
df['b'] = b
df['d'] = d
```

```python
df.Displacement = df.Displacement.str.replace('T','').str.replace('L','').str.replace('纯电动','').str.replace('\'','').str.replace(',','').str.replace(' ','')
ss = df.Displacement
ss_max = ss[-3:]
ss_min = ss[:3]
```

```python
regex = '[0-9]*\.[0-9]+'
ss = df['Displacement'].str.findall(regex)
ss_max = ss.apply(lambda x:float(x[-1]) if len(x)>0 else 0 )
ss_min = ss.apply(lambda x:float(x[0]) if len(x)>0 else 0 )
df['Displacement_max'] = ss_max
df['Displacement_min'] = ss_min
df.sort_values('Displacement_max')
```

# 最贵的是哪款车? 

```python
wan_per = df['Officialprice'][df['Officialprice'].str.contains('万')]
wan_per = wan_per.str.replace('万','')
wan = wan_per.str.split('-')
for i in range(len(wan)):
    if len(wan[i]) == 1:
        df.loc[i,'Officialprice']=float(wan_per[i])
    else:
        df.loc[i,'Officialprice']=float(wan[i][0])+float(wan[i][1])/2
```

```python
df[df['Officialprice'] == df.groupby('Officialprice').count().tail(1).index[0]]
```
