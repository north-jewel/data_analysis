```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
def zong(labels,a,b,c,d,e,f):
    plt.figure(figsize=(9,7))
    
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_color('none')
    plt.grid(axis = 'y') 
    ax.set_axisbelow(True)
    plt.title('9.25-10.25月度违纪对比')
    plt.xticks(b,labels) 
    plt.bar(b,a, alpha = 0.8, label = '未带电脑总数')
    plt.bar(b,c,bottom=a,color = 'orange', alpha = 0.8, label = '未带手机总数')
    plt.bar(b,d,bottom=a+c,color = 'gray', alpha = 0.8,label = '未关投影次数')
    plt.bar(b,e,bottom=a+c+d,color ='yellow', alpha = 0.8, label = '未关门窗次数')
    plt.bar(b,f,bottom=a+c+d+e,color = 'red', alpha = 0.8, label = '校查违纪')
    plt.legend(bbox_to_anchor=(-0.03, -0.2, 1.05,.1 ),loc=2,ncol=5, mode='expand', borderaxespad=0.5)
    plt.show()
    
b = np.arange(10)
a = np.random.randint(0,100,10)
c = np.random.randint(5,15,10)
d = np.random.randint(2,8,10)
e = np.random.randint(2,5,10)
f = np.random.randint(10,15,10)
labels = ['1802C','1803c','1804C','1805C','1806C1','1806C2','1807C1','1807C2','1808C1','1808C2']
zong(labels,a,b,c,d,e,f)
```
