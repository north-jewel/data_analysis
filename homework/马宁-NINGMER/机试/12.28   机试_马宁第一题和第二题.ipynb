# 第一题

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
def may():
    
    x = np.arange(5)
    money = [39, 8,0, 1, 6]
    
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.set_ylim([0, 50])
    plt.grid(axis = 'y')
    ax.set_axisbelow(True)
    plt.bar(x, money)
    plt.title('1802C班违纪情况')
    plt.xticks(x, ('未带电脑总数', '未带手机总数', '未关投影仪次数', '未关门窗次数','校查违纪'))
    for a,b in zip(x,money):
        plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=15)
    plt.show()

may()
```

# 第二题

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
def zong():
    labels = ['1802C','1803c','1804C','1805C','1806C1','1806C2','1807C1','1807C2','1808C1','1808C2']
    a = np.random.randint(0,300,10)
    c = np.random.randint(10,15,10)
    d = np.random.randint(10,20,10)
    e = np.random.randint(10,25,10)
    f = np.random.randint(10,30,10)
    b = np.arange(10)
    
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_color('none')
    plt.grid(axis = 'y')
    ax.set_axisbelow(True)
    plt.title('9.25-10.25月度违纪对比')
    plt.xticks(b,labels) 
    plt.bar(b,a, alpha = 0.8, label = '未带电脑总数')
    plt.bar(b,c,bottom=a,color = 'orange', alpha = 0.8, label = '未带手机总数')
    plt.bar(b,d,bottom=a+c,color = 'gray', alpha = 0.8,label = '未关投影次数')
    plt.bar(b,e,bottom=a+c+d,color ='yellow', alpha = 0.8, label = '未关门窗次数')
    plt.bar(b,f,bottom=a+c+d+e,color = 'red', alpha = 0.8, label = '校查违纪')
    plt.legend(loc='upper center', ncol=5)
    plt.show()
zong()
```
