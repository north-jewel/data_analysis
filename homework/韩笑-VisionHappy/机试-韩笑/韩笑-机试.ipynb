```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
x = np.arange(5)
height =[39,8,0,1,6]
labels= '未带电脑总数', '未带手机总数', '未关投影仪次数', '未关门窗次数','校查违纪'
def zhou():
    
    plt.title('1802C违纪情况')#标题
    plt.bar(x,height)#画条形图
    plt.xticks(x,labels)#改x轴的标签
    ax =  plt.gca()#获取轴
    ax.set_axisbelow(True)
    for i in ax.spines:
        ax.spines[i].set_color('none')
    plt.grid(axis='y')#以轴为基准画格子
    plt.annotate('未带电脑总数,39',xy=(0,40))#加标注，xy是位置
    plt.annotate('未带手机总数,8',xy=(1,10))
    plt.annotate('未关投影总数,0',xy=(2,1))
    plt.annotate('未关门窗次数,1',xy=(3,1))
    plt.annotate('校查违纪,6',xy=(4,7))

    
zhou()
```

```python
z = np.arange(10)
classroom=['1802C','1803C','1804C','1805C','1806C1','1806C2','1807C1','1807C2','1808C1','1808C2']
def zhou():
    np.random.seed(50)
    a = np.random.randint(0,100,10)
    b = np.random.randint(0,100,10)
    c = np.random.randint(0,100,10)
    d = np.random.randint(0,100,10)
    e = np.random.randint(0,100,10)
    print(a,b,c,d,e)
    plt.title('9.25--10.25月度违纪对比')
    plt.xticks(z,classroom)   
    ax = plt.gca()
    ax.set_axisbelow(True)
    for i in ax.spines:
        ax.spines[i].set_color('none')   
    plt.bar(z,a,label='未带电脑总数')
    plt.bar(z,b,bottom=a,color='orange',label='未带手机总数')
    plt.bar(z,c,bottom=a+b,color='gray',label='未关投影次数')
    plt.bar(z,d,bottom=a+b+c,color='yellow',label='未关门窗次数')
    plt.bar(z,e,bottom=a+b+c+d,color='red',label='校查违纪')
    plt.grid(axis='y')
    plt.legend(bbox_to_anchor=(-0.2, -0.2, 1.3,100), loc=3,ncol=5, mode='expand', borderaxespad=0.1)
zhou()
```
