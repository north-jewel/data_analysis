```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号
```

```python
time = [39,8,0,1,6]
m = ['未带电脑总数','未带手机总数','未关投影仪次数','未关门窗次数','校查违纪']
n = np.arange(5)

# 绘图
plt.bar(n, time, align = 'center',color='steelblue', alpha = 0.8)

ax = plt.gca()
ax.spines['right'].set_color('none')     
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
plt.title('1802C班违纪情况')
# 添加刻度标签
plt.xticks(n,m)
# 设置Y轴的刻度范围
plt.ylim([0,45])
# 为每个条形图添加数值标签    enumerate可以同时获得索引和值
plt.grid(axis='y')
for x,y in enumerate(time):
    plt.text(x,y+1,'%s' %round(y,1),ha='center')
    
plt.show()
```

```python
n = np.arange(10)
m = ['1802C','1803C','1804C','1805C','1806C1','1806C2','1807C1','1807C2','1808C1','1808C2']
labels= ['未带电脑总数', '未带手机总数', '未关投影仪次数', '未关门窗次数','校查违纪']
def compare():
    np.random.seed(100)
    a = np.random.randint(0,300,10)
    b = np.random.randint(0,50,10)
    c = np.random.randint(0,50,10)
    d = np.random.randint(0,50,10)
    e = np.random.randint(0,50,10)
    df = pd.DataFrame([a,b,c,d,e],index=labels)
    plt.title('9.25--10.25月度违纪对比')
    plt.xticks(n,m)   
    ax = plt.gca()  
    ax.spines['left'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    plt.bar(n,a,label='未带电脑总数')
    plt.bar(n,b,bottom=a,color='orange',label='未带手机总数')
    plt.bar(n,c,bottom=a,color='gray',label='未关投影次数')
    plt.bar(n,d,bottom=a,color='yellow',label='未关门窗次数')
    plt.bar(n,e,bottom=a,color='red',label='校查违纪')
    plt.grid(axis='y')
    plt.legend(bbox_to_anchor=(0., -0.2, 1.3, -1), loc=3,ncol=5, mode='expand', borderaxespad=0.)
compare()   
```
