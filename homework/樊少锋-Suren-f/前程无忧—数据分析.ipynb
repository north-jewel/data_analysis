```python
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pylab import *
mpl.rcParams['font.sans-serif']=['SimHei']
```

```python
data = pd.read_csv('qiancheng_data.csv',encoding='gbk')
data.dropna()
```

```python
x = data['company_site'].value_counts().head(10)
x
```

```python
y = x.index
q = x.values
plt.figure(figsize=(20, 6.5))
plt.bar(range(len(y)),q,color='rgb',tick_label=y)
plt.show()
```
