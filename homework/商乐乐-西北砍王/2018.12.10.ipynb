```python
def qiancheng():
    df = pd.read_csv(r'C:\Users\lenovo\Desktop\qiancheng_data.csv',encoding='gbk')
    df1 = df[~df.isin(['北京'])]
    df2 = df1.company_site.str.replace('北京-','')
    

    list = ['顺义区','朝阳区','昌平区','石景山区','海淀区','顺义区','东城区','西城区','丰台区','延庆区','大兴区','房山区']
    df4 = df2[df2.isin(list)]
    df5 = df.iloc[df4.index]
    df5.to_csv('diqu.csv')
    df3=pd.read_csv('diqu.csv') 
    size_sa = df3.groupby('salary')['company_site'].sum() 
    size_sa = df5.groupby('salary')['company_site'].sum()   
    print(size_sa)
qiancheng()
```
