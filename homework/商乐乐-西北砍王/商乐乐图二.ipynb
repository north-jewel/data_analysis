```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
def tu():
    x = [39,8,0,1,6]
    y = range(len(x))
    plt.figure()
    plt.title('1802c班违纪情况')
    plt.xticks(y,['未带电脑总数','未带手机总数','未关投影次数','未关门窗次数','校查违纪'])
    plt.annotate('未带电脑总数,39',xy=(0,40))
    plt.annotate('未带手机总数,8',xy=(1,10))
    plt.annotate('未关投影总数,0',xy=(2,1))
    plt.annotate('未关门窗次数,1',xy=(3,1))
    plt.annotate('校查违纪,6',xy=(4,7))
    plt.grid(axis='y')
    ax = plt.gca()
    ax.spines['left'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    plt.bar(y,x)
tu()
```

```python
z = np.arange(10)
labels= ['未带电脑总数', '未带手机总数', '未关投影仪次数', '未关门窗次数','校查违纪']
classe=['1802C','1803C','1804C','1805C','1806C1','1806C2','1807C1','1807C2','1808C1','1808C2']
def bar():
    np.random.seed(50)
    a = np.random.randint(0,300,10)
    b = np.random.randint(0,50,10)
    c = np.random.randint(0,50,10)
    d = np.random.randint(0,50,10)
    e = np.random.randint(0,50,10)
    df = pd.DataFrame([a,b,c,d,e],index=labels)
    plt.title('9.25--10.25月度违纪对比')
    plt.xticks(z,classroom)   
    ax = plt.gca()  
    ax.spines['left'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    plt.bar(z,a,label='未带电脑总数')
    plt.bar(z,b,bottom=a,color='orange',label='未带手机总数')
    plt.bar(z,c,bottom=a,color='green',label='未关投影次数')
    plt.bar(z,d,bottom=a,color='yellow',label='未关门窗次数')
    plt.bar(z,e,bottom=a,color='red',label='校查违纪')
    plt.grid(axis='y')
    plt.legend(loc='best')
bar()
```
