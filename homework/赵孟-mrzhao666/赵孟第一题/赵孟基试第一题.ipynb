```python
import numpy as np,pandas as pd,matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']
import os
```

```python
class WeiJi:
    def __init__(self,data = [],index = [],title = None):
        self.data = data
        self.index = index
        self.title = title
        
    def dange(self,path = None):
        df = pd.DataFrame(data = self.data,index = self.index,columns = [self.title])
        x_z = range(len(df))
        plt.bar(x = x_z,height = df[self.title],width = 0.5)
        plt.xticks(x_z,self.index)
        plt.title(self.title+'违纪情况',size = 20,color = 'red')
        plt.grid(axis = 'y')
        ax = plt.gca()
        for i in ['right','top','left','bottom']:
            ax.spines[i].set_color('none')
        for i in range(len(self.index)):
            plt.annotate(self.index[i]+str(df.iloc[i][0]),xy=(x_z[i],df.iloc[i][0]),size = 12,color = 'yellow')
        if path != None:
            path1 = r''
            path1 = os.path.join(path,self.title+'违纪情况.png')
            plt.savefig(path1)
        plt.show()
        return plt
```

```python
x = WeiJi(data = [39,8,0,1,6],index = ['未带电脑总数','未带手机总数','未关投影仪次数','未关门次数','校查违纪'],title = '1802C班')
```
