```python
import numpy as np,pandas as pd,matplotlib.pyplot as plt
```

```python
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']   #中文
mpl.rcParams['axes.unicode_minus']=False  
```

```python
def yuedu(data = {},index = ['未带电脑总数','未带手机总数','未关投影仪次数','未关门次数','校查违纪'],title):
    df = pd.DataFrame(data = data,index = index)
    list1 = range(len(df))
    color_list = ['blue','orange','gray','yellow','red']
    plt.title(title)
    cum = df.cumsum()
    plt.legend()
    ax = plt.gca()
    for i in ['right','top','left','bottom']:
                ax.spines[i].set_color('none')
    bottom = None
    for i in list1[::-1]:
        plt.bar(x = list1,height = cum.iloc[i],color = color_list[i],bottom = bottom,label = index[i])
    plt.legend()
    plt.grid(axis = 'y')
    plt.xticks(range(5),df.columns)
    plt.legend(bbox_to_anchor=(0., -0.2, 1.5, -1), loc=3,ncol=5, mode='expand', borderaxespad=0.)
    path1 = r'C:/Users/赵孟/Desktop/月度违纪.png'
    plt.savefig(path1)
    plt.show()
```

```python
data = {'1802C':[10,36,20,59,74],
        '1803C':[94,36,20,59,74],
        '1804C':[10,36,20,59,98],
        '1805C':[10,36,20,59,64],
        '1806C':[10,36,20,59,86]}
yuedu(data = data,title = '9.25-10.25月度违纪对比')
```
