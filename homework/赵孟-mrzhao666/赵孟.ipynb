# 最贵的是哪款车?

```python
def new_int(data_list):
    return_list = [float(i) for i in data_list]
    return return_list

def a(df,off):
    data = df[off]
    data = data.replace('万','').split('-')
    data = new_int(data)
    return sum(data)/len(data)
```

```python
df.Officialprice = df.apply(a,axis = 1,off = 'Officialprice')
```

```python
df.nlargest(1,'Officialprice')
```
