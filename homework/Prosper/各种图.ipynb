```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-np.pi,np.pi,256,endpoint=True)
```

```python
c,s=np.cos(x),np.sin(x)
```

```python
plt.figure(1)
```

```python
plt.plot(x,s,'r--',label='sin')
plt.plot(x,c,'g--',label='cos')
plt.plot(x,np.abs(x)<0.5)

ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

ax.xaxis.set_ticks_position('bottom')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$'])

ax.yaxis.set_ticks_position('left')
plt.yticks(np.linspace(-1,1,5))

for label in ax.get_xticklabels()+ax.get_yticklabels():
    label.set_fontsize(10)
    label.set_bbox(dict(facecolor='m',edgecolor='red',alpha=0.2))
ax.grid(color='y')

plt.fill_between(x,np.abs(x)<0.5,c,c>0.1,color='blue')

t=1
plt.plot([t,t],[0,np.cos(t)],'y',linewidth=3,linestyle='--')
plt.annotate('y=cos(1)',xy=(t,np.cos(1)),xycoords='data',xytext=(+10,+20),
            textcoords='offset points',arrowprops=dict(arrowstyle='->',
            connectionstyle='arc3,rad=.8'))

plt.legend(loc=1)

plt.show()
```

# scatter 散点图

```python
fig=plt.figure()
#ax=fig.add_subplot(331)  ax和plt就一样
n=128
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)
T=np.arctan2(X,Y)#上色

plt.axes([0.025,0.025,0.95,0.95],facecolor='m')#left ,bottom, width, height

#plt.scatter(X,Y,s=75,c='#FFOOFF',alpha=.5)
plt.scatter(X,Y,s=175,c=T,alpha=.5)#RGB
plt.xlim(-1.5,1.5)#x的范围
#plt.xticks([]) #x轴的标记
plt.ylim(-1.5,1.5)#y的取值范围
#plt.yticks([])
plt.axis()
plt.title('scatter')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 柱状图

```python
fig.add_subplot(332)
n=10
X=np.arange(n)
Y1=(1-X/float(n)*np.random.uniform(0.5,1.0,n))
Y2=(1-X/float(n)*np.random.uniform(0.5,1.0,n))

plt.bar(X,Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='w')

for x,y in zip(X,Y1):
    plt.text(x+0.1,y+0.02,'%.2f'%y,ha='center',va='bottom')#添加注释 va=bottom 是字的底部
for x,y in zip(X,Y2):
    plt.text(x+0.1,-y-0.02,'%.2f'%y,ha='center',va='top')

plt.show()
```

```python
def f(x,y):
    return (1-x/2 + x**5+y**3)*np.exp(-x**2-y**2)
n=256
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
X,Y=np.meshgrid(x,y)
plt.contourf(X,Y,f(X,Y),8,alpha=.75,cmap=plt.cm.hot)

plt.show()
```
