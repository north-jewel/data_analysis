```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                          "bar", "bee", "bar", "bee"],
                    "B": ["one", "one", "one", "two", "two",
                      "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],'E':[7,8,4,5,12,1,2,6,4]})
```

```python
df
```

```python
pd.pivot_table(df,index=['A','C'])
```

```python
pd.pivot_table(df,index=['A','C'],aggfunc=np.max)
```

```python
pd.pivot_table(df,index=['A','C'],values=['E'])
```

```python
pd.pivot_table(df,index=['A','C'],values=['E'],columns=['B'])
```

```python
pd.pivot_table(df,index=['A','C'],values=['E'],columns=['B'],aggfunc=np.max)
```

```python
pd.pivot_table(df,index=['A','C'],columns=['B'],fill_value=5)
```

```python
pd.pivot_table(df,index=['A'],columns=['B'],values=['E'],aggfunc=np.sum)
```

```python
pd.pivot_table(df,index=['A'],columns=['B'],values=['E'],aggfunc=np.min)
```

```python
pd.pivot_table(df,columns=['B'],values=['E'],aggfunc=[np.max,len])
```
